# Basic proof of concept for psy-vtk and UGRID files

This notebook just tests the visualization of geo-referenced data using psy-vtk with a UGRID-file (see [psy-maps example on UGRID-files](https://psyplot.readthedocs.io/projects/psy-maps/en/latest/examples/example_ugrid.html#visualizing-ugrid-data) if you want to see how it looks with matplotlib)

```python
import psyplot.project as psy
import psy_vtk.plotters as pvtk
```

```python
%matplotlib inline
```

```python
ds = psy.open_dataset('ugrid_demo.nc')
ds
```

```python
data = ds.psy.Mesh2_height.psy[0]
```

```python
globe_plot = pvtk.GlobePlotter(data, cmap='Blues')
globe_plot.ax.camera_position = [-34855191039.83234, -7270697895.145699, 15100163601.493084]
disp = globe_plot.ax.show()
disp
```

```python
globe_plot.update(datagrid=True)
globe_plot.ax.show()
```
